Tener columnas con varianza cero en tu dataset es un problema porque esas columnas no aportan ninguna información útil al análisis ni al modelo de machine learning. Aquí te explico por qué:

# Problemas de las columnas con varianza cero
1. No aportan diferenciación
Una columna con varianza cero tiene el mismo valor en todas las filas, por lo que no ayuda a distinguir entre clases o predecir una variable objetivo.

Ejemplo: Si la columna usaapp = 1 para todos los registros, entonces no sirve para nada en un modelo.

2. Ruido innecesario
Aunque no afecten directamente la predicción, agregan dimensionalidad sin aportar valor, lo que puede confundir o ralentizar los algoritmos.

3. Problemas con algoritmos sensibles a la varianza
Modelos como SVM, PCA, regresión logística, KNN o árboles pueden verse afectados en precisión o desempeño computacional.

Algunos procesos de escalamiento (como StandardScaler) usan la desviación estándar; si es cero, se puede producir una división por cero.

4. Tiempo y recursos malgastados
Si no limpias esas columnas, estás alimentando al modelo con datos inútiles, lo que aumenta innecesariamente el tiempo de entrenamiento.

## La kurtosis ##  

es una medida estadística que describe la forma de la distribución de una variable, específicamente:
Qué tan "afilada" o "aplanada" es la curva en comparación con una distribución normal.

Interpretación de los valores:
-Kurtosis ≈ 3 (o 0 si es "exceso de kurtosis") → distribución normal.

-Kurtosis > 3 → leptocúrtica:
        Distribución con colas más pesadas y pico más alto.
        Implica mayor presencia de outliers.

-Kurtosis < 3 → platicúrtica:
        Distribución más "plana", con colas más ligeras.
        Menor probabilidad de valores extremos.

¿Para qué sirve en análisis de datos?
Detectar riesgo: en finanzas, alta kurtosis implica mayor riesgo por eventos extremos.

Preparación de datos: si una variable tiene una kurtosis muy alta, puede ser bueno transformarla (log, raíz, etc.) para suavizarla.

la kurtosis puede ayudarte a decidir si debes aplicar transformaciones.

## Un Q-Q plot es una herramienta gráfica que compara la distribución de tus datos reales con una distribución teórica (en este caso, la normal).

¿Qué muestra?
En el eje X: los cuantiles teóricos de la distribución normal.

En el eje Y: los cuantiles reales de tus datos (df['total']).

¿Cómo interpretarlo?
Si los puntos caen sobre la línea recta → tus datos siguen una distribución normal.

Si los puntos se curvan hacia arriba o abajo:
 
    Curva en forma de S → datos asimétricos.
    
    Cola arriba o abajo muy lejos → presencia de outliers o colas pesadas.

Ejemplos visuales:
✔️ Línea recta perfecta → buena aproximación a normal.

❌ Curva hacia arriba en los extremos → distribución con colas pesadas (alta kurtosis).

❌ Curva en forma de S → distribución asimétrica (skewed).

```python
from scipy.stats import shapiro

stat, p = shapiro(df['total'])
print("Shapiro-Wilk test: p-value =", p)
```

```python
Q1 = df['total'].quantile(0.25)
Q3 = df['total'].quantile(0.75)
IQR = Q3 - Q1

# Límites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identificar outliers
outliers = df[(df['total'] < limite_inferior) | (df['total'] > limite_superior)]

print(f"Se encontraron {len(outliers)} outliers.")
```

```python
from scipy.stats import zscore

df['z_total'] = zscore(df['total'])

# Outliers con z-score > 3 o < -3
outliers_z = df[(df['z_total'] > 3) | (df['z_total'] < -3)]

print(f"Se encontraron {len(outliers_z)} outliers con Z-score.")
```
